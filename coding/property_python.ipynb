`property` in Python

```python
class Dog:
    def __init__(self, weight):
        self.weight = weight
```

```python
dog = Dog(5)
dog.weight
```

Someone can create a dog with negative weight, not good.

```python
dog = Dog(-5)
dog.weight
```

You can use a setter to prevent people from creating negative-weight dogs.

```python
class Dog:
    def __init__(self, weight):
        self.set_weight(weight)
    
    def set_weight(self, weight):
        if weight <= 0:
            raise ValueError("Weight must be positive!")
        else:
            self.weight = weight
```

```python
dog = Dog(5)
```

```python
dog = Dog(-5)
```

But you can't prevent people from setting good dogs' weight negative. 

```python
dog = Dog(5)
dog.weight = -5
dog.weight
```

For this you can make use `property`.

```python
class Celsius:
    def __init__(self, temperature = 0):
        self.temperature = temperature

    def to_fahrenheit(self):
        return (self.temperature * 1.8) + 32

    def get_temperature(self):
        print("Getting value")
        return self._temperature

    def set_temperature(self, value):
        if value < -273:
            raise ValueError("Temperature below -273 is not possible")
        print("Setting value")
        self._temperature = value

    temperature = property(get_temperature,set_temperature)
```

```python
class Dog:
    def __init__(self, weight):
        self.set_weight(weight)
    
    def set_weight(self, weight):
        if weight <= 0:
            raise ValueError("Weight must be positive!")
        self.weight = weight
    weight = property(fset=set_weight)
```

```python
dog = Dog(-5)
```

```python
dog = Dog(5)
dog.weight = -5
```

As syntatic sugar, you can use the `@property` decorator.

```python
class Dog:
    def __init__(self, weight):
        self._weight = weight
        
    @property
    def weight(self):
        print("Get weight")
        return self._weight

    @weight.setter
    def weight(self, value):
        print("Set weight")
        if value <= 0:
            raise ValueError("Weight must be positive!")
        self._weight = value    
```

```python
dog = Dog(5)
```

```python
dog.weight
```

```python
dog.weight = 5
```

```python
dog.weight = -5
```

However this does not prevent **creating** negative-weight dogs...

```python
dog = Dog(-5)
dog.weight
```

```python
dog.weight = -5
```
