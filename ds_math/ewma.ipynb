```python
ts = pd.Series(np.random.randint(100, size=100))
```

```python
def ewma(ts, alpha=.5):
    y = np.empty_like(ts)
    y[0] = ts[0]
    for i in range(1, len(y)):
        y[i] = alpha*ts[i] + (1-alpha)*y[i-1]
    return y
```

```python
plt.figure(figsize=(15, 8))
plt.plot(ts, '--.')
plt.plot(ewma(ts, alpha=0.1), '-.', label='alpha=0.1')
plt.plot(ewma(ts, alpha=0.3), '-.', label='alpha=0.3')
plt.plot(ewma(ts, alpha=0.5), '-.', label='alpha=0.5')
plt.legend()
plt.show()
```
